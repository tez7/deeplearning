```python
from ensure import ensure_annotations
```

```python
def multi_numbers(a: int, b: int) -> int:
    return a * b
print(multi_numbers(a = 3, b =2))
print(multi_numbers(a = 3, b ="2"))

# despite i gave input type as int for b it accepted string also and concatenated it. to avoid this we use ensure_annotations
```

```python
@ensure_annotations
def multi_numbers(a: int, b: int) -> int:
    return a * b
print(multi_numbers(a = 3, b =2))
print(multi_numbers(a = 3, b ="2"))
```
